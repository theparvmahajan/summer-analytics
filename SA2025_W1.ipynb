### 1. What is name of car that has the highest horsepower?  

```python
#your code here
max_hp_idx = df['horsepower'].idxmax()
car_highest_hp = df.loc[max_hp_idx, 'name']
max_hp_value = df.loc[max_hp_idx, 'horsepower']
print(f"\nCar with highest horsepower: {car_highest_hp}")
print(f"   Horsepower: {max_hp_value}")

#Car with highest horsepower: pontiac grand prix
#Horsepower: 230.0
```

### 2. How many cars have mpg ≥ 35?

```python
#your code here
cars_mpg_35_plus = df[df['mpg'] >= 35]
count_mpg_35_plus = len(cars_mpg_35_plus)
print(f"\nNumber of cars with MPG ≥ 35: {count_mpg_35_plus}")

#Number of cars with MPG ≥ 35: 36
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000) & (df['horsepower'].notna())]
most_common_origin = filtered_cars['origin'].mode()[0]
origin_counts = filtered_cars['origin'].value_counts()
print(f"\nMost common origin for cars with HP > 100 and weight < 3000: {most_common_origin}")
print(f"   Origin distribution: {origin_counts.to_dict()}")

#Most common origin for cars with HP > 100 and weight < 3000: usa
#Origin distribution: {'usa': 9, 'europe': 8, 'japan': 6}
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
japan_cars = df[df['origin'] == 'japan']
mean_acceleration_japan = round(japan_cars['acceleration'].mean(), 2)
print(f"\nMean acceleration of cars from Japan: {mean_acceleration_japan}")

#Mean acceleration of cars from Japan: 16.17
```

### 5. Which year had the highest average mpg?

```python
#your code here
df['full_year'] = df['model_year'].apply(lambda x: 1900 + x if x >= 70 else 2000 + x)
avg_mpg_by_year = df.groupby('full_year')['mpg'].mean()
year_highest_avg_mpg = avg_mpg_by_year.idxmax()
highest_avg_mpg = avg_mpg_by_year.max()
print(f"\nYear with highest average MPG: {year_highest_avg_mpg}")
print(f"   Average MPG: {highest_avg_mpg:.2f}")

#Year with highest average MPG: 1980
#Average MPG: 33.70
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg_cars = df[df['mpg'] > median_mpg].copy()
above_median_mpg_cars['hp_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

best_ratio_idx = above_median_mpg_cars['hp_weight_ratio'].idxmax()
best_ratio_car = df.loc[best_ratio_idx, 'name']
best_ratio_value = above_median_mpg_cars.loc[best_ratio_idx, 'hp_weight_ratio']

print(f"\nCar with best HP/weight ratio among above-median MPG cars:")
print(f"   Car: {best_ratio_car}")
print(f"   HP/Weight ratio: {best_ratio_value:.6f}")
print(f"   MPG: {df.loc[best_ratio_idx, 'mpg']}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(12, 8))
avg_mpg_by_year_origin = df.groupby(['full_year', 'origin'])['mpg'].mean().unstack()

for origin in avg_mpg_by_year_origin.columns:
    plt.plot(avg_mpg_by_year_origin.index, avg_mpg_by_year_origin[origin], 
             marker='o', linewidth=2, label=origin.capitalize())

plt.title('Evolution of Average MPG Over Years by Origin', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.legend(title='Origin', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nMulti-line plot created showing MPG evolution by origin over years")
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
plt.figure(figsize=(12, 8))

df_plot = df[df['horsepower'].notna()]

sns.scatterplot(data=df_plot, x='horsepower', y='weight', 
                hue='origin', size='mpg', sizes=(50, 200),
                hue_order=['japan', 'europe', 'usa'], alpha=0.7)

plt.title('Car Weight vs Horsepower by Origin (Size = MPG)', fontsize=16, fontweight='bold')
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("\nScatterplot created showing Weight vs Horsepower relationship")
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
car_stats = df.groupby('name').agg({
    'model_year': 'count',  # Number of appearances
    'mpg': ['mean', 'std']   # Mean and std of mpg
}).round(2)


car_stats.columns = ['appearances', 'avg_mpg', 'mpg_std']

consistent_cars = car_stats[
    (car_stats['appearances'] > 1) & 
    (car_stats['mpg_std'] < 1.0) &
    (car_stats['mpg_std'].notna()) 
]

consistent_cars_sorted = consistent_cars.sort_values(['appearances', 'avg_mpg'], ascending=[False, False])

print(f"\nMost consistent car models (std(mpg) < 1.0, multiple years):")
print(f"Found {len(consistent_cars_sorted)} consistent car models:")
for name, row in consistent_cars_sorted.head(10).iterrows():  # Show top 10
    print(f"   {name}: {int(row['appearances'])} appearances, "
          f"avg MPG: {row['avg_mpg']:.2f}, std: {row['mpg_std']:.3f}")

print(f"\nMost consistent car: {consistent_cars_sorted.index[0] if len(consistent_cars_sorted) > 0 else 'None found'}")

```
